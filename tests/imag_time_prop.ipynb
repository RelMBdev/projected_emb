```python
import sys
import os
sys.path.insert(0,'../FnT')
from it_tddft_class import initialize, mainrun
```

```python
#set options

basis = 'def2-svp'
func = 'hf'
tstep = 1.0
loewdin = False
debug = False
mem = 2
maxit = 25

geom_data = ' O -1.4626 0.0000 0.0000\n \
         H -1.7312 0.9302 0.0000\n \
         H -0.4844 0.0275 0.0000\n \
         symmetry c1\n \
         no_com\n \
         no_reorient\n'

wfn_mol = initialize(geom_data,basis)
ene_list ,dip_list = mainrun(wfn_mol, func, tstep, maxit, loewdin,debug, mem)
                                                                                            
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
n = np.linspace(1,maxit,maxit)
plt.plot(n, np.array(dip_list))
plt.title("dipole moment / iterations")
plt.xlabel("n.iter")
plt.ylabel("\u03BC (t)")
plt.show()
```

```python
plt.plot(n[10:], np.array(ene_list)[10:])
plt.title("\u0394 E / iterations")
plt.xlabel("n.iter")
plt.ylabel("\u0394 E")
plt.show()
```
